# Random Forest : Ensemble Learning Modeli (birden fazla modelden oluşan büyük eğitme modelleri) (random forest, decision tree'lerin bir araya gelerek oluşmasıyla ortaya çıkar)

## Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("tumor_dataset.csv")
df.head(7)
```

```python
df.drop(["id","Unnamed: 32"] , axis = 1 , inplace = True)
df["diagnosis"] = [1 if each == "M" else 0 for each in df["diagnosis"]]
```

```python
df.describe()
```

# x-y & normalization

```python
y = df["diagnosis"]
x_data = df.drop(["diagnosis"], axis = 1)
x = (x_data - x_data.min()) / (x_data.max() - x_data.min())
x.describe()
```

# Train-test

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
x_train.info()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state = 42)
model.fit(x_train , y_train)
```

## Decision Tree Predict

```python
print("Decision Tree accuracy : ", model.score(x_test , y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators = 100 , random_state = 42)
modelRF.fit(x_train , y_train)
```

## Random Forest Predict

```python
print("Random Forest accuracy : ", modelRF.score(x_test , y_test))
```

# Accuracy comparison

```python
print("Decision Tree accuracy : ", model.score(x_test , y_test))
print("Random Forest accuracy : ", modelRF.score(x_test , y_test))
```
