## Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("tumor_dataset.csv")
df.head(7)
```

```python
df.drop(["id","Unnamed: 32"] , axis = 1 , inplace = True)
df["diagnosis"] = [1 if each == "M" else 0 for each in df["diagnosis"]]
df.describe()
```

## x-y & normalization

```python
y = df["diagnosis"]
x_data = df.drop(["diagnosis"], axis = 1)
x = (x_data - x_data.min()) / (x_data.max() - x_data.min())
x.describe()
```

# Train-test

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

# We are trying the Random Forest model for Confusion Matrix

```python
from sklearn.ensemble import RandomForestClassifier
modelRF = RandomForestClassifier(n_estimators = 100 , random_state = 42)
modelRF.fit(x_train , y_train)
```

```python
print("Random Forest accuracy : ", modelRF.score(x_test , y_test))
```

# Confusion Matrix

```python
y_prediction = modelRF.predict(x_test)
y_true = y_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true , y_prediction)
```

## visualization

```python
import seaborn as sns
```

```python
f, ax = plt.subplots(figsize = (5,5))
sns.heatmap(cm,annot = True , linewidth = 0.5 , linecolor = "red" , fmt = ".0f", ax=ax)
plt.xlabel("y_prediction (y_head)")
plt.ylabel("y_true (y_test)")
```
