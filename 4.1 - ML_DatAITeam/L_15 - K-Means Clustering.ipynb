# K-Means algoritmasında bir K değeri belirlenir. Dataset belirlenen bu değer kadar gruba bölünür.

#

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## Creating Dataset

```python
#class1
x1 = np.random.normal(25,5,1000) # 25'e maksimum 5 uzaklıkta 1000 adet rastgele değer oluştur
y1 = np.random.normal(25,5,1000)

#class2
x2 = np.random.normal(55,5,1000)
y2 = np.random.normal(60,5,1000)

#class1
x3 = np.random.normal(55,5,1000)
y3 = np.random.normal(15,5,1000)
```

```python
x = np.concatenate((x1,x2,x3), axis = 0)
y = np.concatenate((y1,y2,y3), axis = 0)
```

```python
dictionary = {"x" : x , "y" : y}
```

```python
data = pd.DataFrame(dictionary)
```

```python
plt.scatter(x1,y1, color="blue", alpha = 0.1)
plt.scatter(x2,y2, color="red", alpha = 0.1)
plt.scatter(x3,y3, color="green", alpha = 0.1)
plt.xlabel("x")
plt.ylabel("y")
```

# K-Means

```python
from sklearn.cluster import KMeans
wcss = []

#bir döngüde tüm k değerlerini deniyoruz.
for k in range(1,15):
    kmeans = KMeans(n_clusters=k , n_init=1)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_) #inertia her k değeri için wcss değerini bulur

plt.plot(range(1,15), wcss)
plt.xlabel("k (cluster value)")
plt.ylabel("wcss")
```

## Dirsek yaptığı nokta(3) en uygun k değeridir.

```python
model = KMeans(n_clusters=3 , n_init=10)
clusters = model.fit_predict(data)
data["label"] = clusters
```

```python
plt.scatter(data.x[data.label==0],data.y[data.label==0], color="blue", alpha = 0.3)
plt.scatter(data.x[data.label==1],data.y[data.label==1], color="green", alpha = 0.3)
plt.scatter(data.x[data.label==2],data.y[data.label==2], color="red", alpha = 0.3)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color="black", label =  "clusters")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
```
