```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Creating Dataset

```python
#class1
x1 = np.random.normal(25,5,1000) # 25'e maksimum 5 uzaklıkta 1000 adet rastgele değer oluştur
y1 = np.random.normal(25,5,1000)

#class2
x2 = np.random.normal(55,5,1000)
y2 = np.random.normal(60,5,1000)

#class1
x3 = np.random.normal(55,5,1000)
y3 = np.random.normal(15,5,1000)

x = np.concatenate((x1,x2,x3), axis = 0)
y = np.concatenate((y1,y2,y3), axis = 0)

dictionary = {"x" : x , "y" : y}
data = pd.DataFrame(dictionary)
```

```python
plt.scatter(x1,y1, color="black", alpha = 0.1)
plt.scatter(x2,y2, color="black", alpha = 0.1)
plt.scatter(x3,y3, color="black", alpha = 0.1)
plt.xlabel("x")
plt.ylabel("y")
```

# Dendrogram

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
merg = linkage(data, method = "ward")
dendrogram(merg, leaf_rotation = 90)
plt.xlabel("data points")
plt.ylabel("euclidean distance")
```

# Hierarchical Clustering HC

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters = 3 , affinity = "euclidean", linkage = "ward")
cluster = hierarchical_cluster.fit_predict(data)
```

```python
data["label"] = cluster
```

```python
plt.scatter(data.x[data.label == 0],data.y[data.label == 0], color="green", alpha = 0.3)
plt.scatter(data.x[data.label == 1],data.y[data.label == 1], color="blue", alpha = 0.3)
plt.scatter(data.x[data.label == 2],data.y[data.label == 2], color="red", alpha = 0.3)
plt.xlabel("x")
plt.ylabel("y")
```
