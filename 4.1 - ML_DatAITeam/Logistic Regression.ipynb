# Kullanmayacağım sütunları dropluyorum

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace = True)
```

```python
data.info()
```

# diagnosis (kanserin iyi ya da kötü huylu olma durumu) object olduğu için kullanamıyoruz. Bu yüzden int olarak belirleyip 1 ya da 0 olarak ayarlıyorumç

```python
data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
data.diagnosis
```

```python
data.info()
```

## Dataset'imin temizliği ile ilgili aşamayı tamamladım.

```python
y = data.diagnosis.values.reshape(-1,1)
x_ham_veri = data.drop(["diagnosis"],axis=1)
```

```python
y
```

```python
x_ham_veri
```

## Tüm featureları 0 ve 1 aralığına getirmek için normalization yapıyorum. Bu işlemi yaparsam çok yüksek değerli featurelar, diğer featurları baskılayabilir.

```python
x = (x_ham_veri - x_ham_veri.min()) / (x_ham_veri.max() - x_ham_veri.min()).values
x
```

# Train - test split aşaması

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,  y_test = train_test_split(x,y,test_size = 0.2, random_state=42)
```
