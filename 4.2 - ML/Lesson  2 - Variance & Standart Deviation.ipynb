#### Standart Sapma, bir veri setindeki veriler arasındaki sapmayı verir. Standart Sapma ne kadar yüksekse, veriler arasında yüksek farklar olduğunu anlarız

##### Varyans, verilerin aritmetik ortalamadan sapmalarının karelerinin toplamının örnek sayısının 1 eksiğine bölümüdür (standart sapmanın karekök alınmamış hâli)

```python
# np.random.normal(sapma yaşanacak değer, standart sapma , veri sayisi)
gelirler = np.random.normal(5000,2000,1000)
gelirler
```

### Variance

```python
gelirler.var()
```
